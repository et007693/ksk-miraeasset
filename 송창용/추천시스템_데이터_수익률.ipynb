## 수익률 계산

```python
# 시간
now = datetime.datetime.now()
now2 = now - datetime.timedelta(days=11)  # 수정된 부분
```

```python
# KRX
df_krx = fdr.StockListing('KRX')
kr_list = df_krx['Code']

rate_list = []

for k in tqdm(kr_list):
    try:
        df = fdr.DataReader(k, now2, now)
        rate = (df['Close'][-1] - df['Close'][0]) / df['Close'][-1]
        rate_list.append(rate)
    except:
        rate_list.append(0)

df_krx['rate'] = rate_list
```

## 추천시스템 데이터 생성

```python
stock_list = custom[['BUY1_ITM_M1_1','BUY1_ITM_M1_2','BUY1_ITM_M1_3','BUY1_ITM_M2_1','BUY1_ITM_M2_2','BUY1_ITM_M2_3','BUY1_ITM_M3_1','BUY1_ITM_M3_2','BUY1_ITM_M3_3','BUY1_ITM_M4_1','BUY1_ITM_M4_2','BUY1_ITM_M4_3']] # 주식 상위 구매 리스트
# 데이터 재구조화
# stock_list.dropna(how='all', inplace=True)
# stock_list = stock_list.reset_index(drop = True).reset_index()
stock_list.reset_index(inplace=True)
stock_list_rc = pd.melt(stock_list, id_vars = 'index').rename(columns = {'value':'Code'});stock_list_rc 
```

```python
# 한국 주식 + 미국 주식
cod_rate = pd.concat([korea[['Code','rate']], america[['Code','rate']]])
cod_rate.drop_duplicates(inplace=True)
cod_rate =cod_rate.fillna(0)
# 중복된 'Code' 값 중에서 'rate' 열의 값이 0인 경우 삭제
cod_rate = cod_rate[~((cod_rate['rate'] == 0) & (cod_rate.duplicated('Code') | cod_rate.duplicated('Code', keep='last')))];cod_rate
```

```python
# 수익률 결합
code_rate = pd.merge(stock_list_rc, cod_rate, on = 'Code', how = 'left');stock_list_rc
```

```python
code_rate2 = code_rate.sort_values(by=['index','variable','Code'])
code_rate3 = code_rate2.dropna(subset='Code').reset_index(drop=True)
code_rate4 = code_rate3[['index','Code','rate']].fillna(0);code_rate4
```
