### 월 말일 기준 평가액 1위 국내주식 [종목번호, 거래소, 평가액, 매입액] 삭제
#### DMST_AST1_ITM, DMST_AST1_MKT, DMST_AST1_EVAL, DMST_AST1_PCHS

```python
for i in range(1,4):
    print("지우기 전 column 수 : ", custom.shape[1])
    columns_to_remove = [col for col in custom.columns if col.startswith('DMST_AST' + str(i))]
    custom = custom.drop(columns=columns_to_remove)
    print("지운 후 column 수 : ", custom.shape[1])
```

### 특정 기간 동안의 주식 매수액 1위 [종목번호, 거래소, 매수액] 삭제
#### BUY1_ITM, BUY1_MKT, BUY1_AMT

```python
for i in range(1,4):
    print("지우기 전 column 수 : ", custom.shape[1])
    columns_to_remove = [col for col in custom.columns if col.startswith('BUY' + str(i))]
    custom = custom.drop(columns=columns_to_remove)
    print("지운 후 column 수 : ", custom.shape[1])
```

### 특정 기간 동안의 주식 매도액 1위 [종목번호, 거래소, 매수액] 삭제
#### SEL1_ITM, SEL1_MKT, SEL1_AMT

```python
for i in range(1,4):
    print("지우기 전 column 수 : ", custom.shape[1])
    columns_to_remove = [col for col in custom.columns if col.startswith('SEL' + str(i))]
    custom = custom.drop(columns=columns_to_remove)
    print("지운 후 column 수 : ", custom.shape[1])
```

### M1_1, M1_2, M1_3로 되어 있는 데이터를 합계로 합침

```python
columns = ['BUY_CNT_M','SEL_CNT_M','BUY_ITM_CNT_M','SEL_ITM_CNT_M','BUY_AMT_M','SEL_AMT_M',
          'STK_IN_M','STK_OUT_M','MNY_IN_M','MNY_OUT_M','CONN_DYS_M','MTS_DYS_M','HTS_DYS_M']
for col in columns:
    for i in range(1,5):
        print(col + str(i) + "지우기 전 column 수 : ", custom.shape[1])
        # 새로운 열 생성
        sum_column = custom.loc[:, custom.columns.str.startswith(col + str(i))].sum(axis=1)
        # 기존 열 삭제
        custom = custom.drop(columns=custom.columns[custom.columns.str.startswith(col + str(i))])
        custom[col + str(i) + '_sum'] = sum_column
        print(col + str(i) + "지운 후 column 수 : ", custom.shape[1])
```
