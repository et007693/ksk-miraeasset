```python
import pandas as pd

custom = pd.read_csv('../data/cs_mkt_dataset/cs_data.csv')
```

```python
pattern= r'BUY1_ITM+'
BUY= custom.filter(regex=pattern) # 열이나 행 선택
BUY
```

```python
etf= pd.read_csv('해외etf_sector.csv')
etf
```

```python
merged= pd.merge(BUY, etf, left_on=BUY.columns[0], right_on='Symbol', how='left')
merged.rename(columns={'Sector':f'sector0'}, inplace=True)
merged.drop(columns=['Name', 'Symbol'], inplace=True)
```

```python
for i in range(len(BUY.columns)):
    merged= pd.merge(merged, etf, left_on=BUY.columns[i], right_on='Symbol', how='left')
    merged.rename(columns={'Sector':f'sector{i+1}'}, inplace=True)
    merged.drop(columns=['Name', 'Symbol'], inplace=True)

merged
```

```python
Trans= merged.T
```

```python
Trans.to_csv('Merged해외ETF.csv', encoding='utf-8')
```
