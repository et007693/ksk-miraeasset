```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time

# 전체 데이터를 담을 리스트 초기화
all_data = []

# 웹 페이지 URL 설정
url = 'https://etfdb.com/etfs/'

# 웹 페이지 내용 가져오기
response = requests.get(url)

# HTTP 요청이 성공적으로 완료되었는지 확인
if response.status_code == 200:
    # BeautifulSoup을 사용하여 HTML 파싱
    soup = BeautifulSoup(response.text, 'html.parser')

    # <div class='transition-wrapper'> 요소를 찾기
    table = soup.find('table',{'id':'sector-b1-table'}).find('tbody')
    trs = table.find_all('tr')
    urls= []
    for tr in trs: 
        sector = tr.find('td', {'class': 'title'})
        pages= tr.find('td', {'class':'number_of_etfs_field'}).text
        a_tag = sector.find('a')
        if a_tag:
            href_value = a_tag.get('href')
            urls.append(['https:'+href_value+'#etfs&sort_name=assets_under_management&sort_order=desc&page=',(int(pages)//24)+1])

for i in range(len(urls)):
    print(i)
    # 여러 페이지를 크롤링하려면 페이지 수만큼 반복
    for j in range(1, urls[i][1]):  # 예시로 1부터 3까지 3페이지를 크롤링합니다.
        driver = webdriver.Chrome('C:\\Users\\kim\\Downloads\\chromedriver-win64\\chromedriver-win64\\chromedriver.exe')

    # 시작 URL 설정
        base_url = urls[i][0]
#         'https://etfdb.com/etfs/sector/consumer-discretionaries/#etfs&sort_name=assets_under_management&sort_order=desc&page='

        url = base_url + str(j)

        # 웹 페이지 열기
        driver.get(url)

        # 페이지 내용을 가져오기 위해 시간 대기 (필요한 경우)
        time.sleep(2)  # 2초 대기 (웹 페이지 로드 완료까지 기다림)

        # 페이지 내용 가져오기
        page_source = driver.page_source

        # BeautifulSoup을 사용하여 HTML 파싱
        soup = BeautifulSoup(page_source, 'html.parser')

        # 테이블에서 데이터 추출
        table = soup.find('table', {'id': 'etfs'})
        trs = table.find_all('tr')
        data_list = []

        for index, tr in enumerate(trs):
            # 첫 번째와 마지막 요소는 제외
            if index == 0 or index == len(trs) - 1:
                continue

            symbol = tr.find('td', {'data-th': 'Symbol'}).text
            name = tr.find('td', {'data-th': 'ETF Name'}).text
            data_list.append([symbol, name, urls[i][0].split('/')[5]])

        # 현재 페이지에서 수집한 데이터를 전체 데이터 리스트에 추가
        all_data.extend(data_list)

        # 웹 드라이버 종료
        driver.quit()

# 모든 페이지에서 수집한 데이터를 데이터프레임으로 변환
import pandas as pd
df = pd.DataFrame(all_data, columns=['Symbol', 'Name', 'Sector'])

# 데이터프레임 출력
print(df)
```

```python
# 데이터프레임을 CSV 파일로 저장
df.to_csv('해외etf_sector.csv', index=False)  # index=False를 사용하여 인덱스를 저장하지 않음
```
