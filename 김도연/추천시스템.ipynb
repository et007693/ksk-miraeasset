```python
import pandas as pd

# 가상의 데이터 생성
data = {
    '사용자': ['사용자1', '사용자2', '사용자3', '사용자4'],
    '주식': [
        {'AAPL', 'GOOGL', 'TSLA'},
        {'AAPL', 'MSFT', 'AMZN'},
        {'TSLA', 'AMZN', 'T', 'DIS'},
        {'AAPL', 'GOOGL', 'MSFT', 'T', 'DIS'},
    ]
}

df = pd.DataFrame(data)
df
```

```python
# 자카드 유사도 계산 함수 정의
def jaccard_similarity(set1, set2):
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union if union > 0 else 0.0

# 추천 함수 정의
def get_recommendations(user, df):
    recommendations = []

    for index, row in df.iterrows():
        if row['사용자'] != user:
            similarity = jaccard_similarity(row['주식'], df[df['사용자'] == user]['주식'].values[0])
            recommendations.append((row['사용자'], similarity))

    recommendations.sort(key=lambda x: x[1], reverse=True)
    return recommendations

# 추천을 받을 사용자 지정
target_user = '사용자1'

# 추천 목록 가져오기
recommendations = get_recommendations(target_user, df)

# 사용자에게 추천된 사용자 및 유사도 출력
print(f"{target_user}에게 추천된 사용자 및 유사도:")
for user, similarity in recommendations:
    print(f"사용자: {user}, 유사도: {similarity}")
```

# 피어슨 상관계수

```python
import pandas as pd

# 가상의 데이터 생성
data = {
    '사용자': ['사용자1', '사용자2', '사용자3', '사용자4'],
    'AAPL': [5, 4, 3, 2],
    'GOOGL': [4, 5, 3, 2],
    'TSLA': [3, 4, 5, 2],
    'MSFT': [4, 3, 2, 5],
    'AMZN': [5, 4, 3, 2],
    'T': [2, 3, 4, 5],
    'DIS': [3, 2, 4, 5],
}

df = pd.DataFrame(data)
df.set_index('사용자', inplace=True)
df
```

```python

# 피어슨 상관계수 계산 함수 정의
def pearson_similarity(user1, user2):
    common_stocks = [stock for stock in user1.index if not pd.isna(user1[stock]) and not pd.isna(user2[stock])]
    if len(common_stocks) == 0:
        return 0.0
    user1_ratings = [user1[stock] for stock in common_stocks]
    user2_ratings = [user2[stock] for stock in common_stocks]
    correlation_coefficient = pd.Series(user1_ratings).corr(pd.Series(user2_ratings))
    return correlation_coefficient

# 추천 함수 정의
def get_recommendations(target_user, df):
    recommendations = []

    for user, row in df.iterrows():
        if user != target_user:
            similarity = pearson_similarity(df.loc[target_user], row)
            recommendations.append((user, similarity))

    recommendations.sort(key=lambda x: x[1], reverse=True)
    return recommendations

# 추천을 받을 사용자 지정
target_user = '사용자1'

# 추천 목록 가져오기
recommendations = get_recommendations(target_user, df)

# 사용자에게 추천된 사용자 및 유사도 출력
print(f"{target_user}에게 추천된 사용자 및 유사도:")
for user, similarity in recommendations:
    print(f"사용자: {user}, 유사도: {similarity}")
```

# 코사인 유사도 v2

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# 가상의 데이터 생성
data = {
    '사용자': ['사용자1', '사용자2', '사용자3', '사용자4'],
    'AAPL': [1, 0, 1, 0],
    'GOOGL': [1, 1, 0, 0],
    'TSLA': [0, 1, 1, 1],
    'MSFT': [0, 1, 0, 0],
    'AMZN': [1, 0, 1, 1],
}

df = pd.DataFrame(data)
df.set_index('사용자', inplace=True)
df
```

```python
# 코사인 유사도 계산 함수 정의
def cosine_similarity_users(user1, user2):
    user1_vector = user1.values.reshape(1, -1)
    user2_vector = user2.values.reshape(1, -1)
    similarity = cosine_similarity(user1_vector, user2_vector)
    return similarity[0][0]

# 추천 함수 정의
def get_recommendations(target_user, df):
    recommendations = []

    for user, row in df.iterrows():
        if user != target_user:
            similarity = cosine_similarity_users(df.loc[target_user], row)
            recommendations.append((user, similarity))

    recommendations.sort(key=lambda x: x[1], reverse=True)
    return recommendations

# 추천을 받을 사용자 지정
target_user = '사용자1'

# 추천 목록 가져오기
recommendations = get_recommendations(target_user, df)

# 사용자에게 추천된 사용자 및 유사도 출력
print(f"{target_user}에게 추천된 사용자 및 유사도:")
for user, similarity in recommendations:
    print(f"사용자: {user}, 유사도: {similarity}")
```

```python
import pandas as pd

# 가상의 구매 이력 데이터 생성 (사용자 ID, ETF, 구매 여부)
data = {
    '사용자 ID': [1, 2, 3, 1, 2],
    'ETF': ['ETF1', 'ETF2', 'ETF1', 'ETF3', 'ETF2'],
    '구매': [1, 1, 1, 1, 1]
}

df = pd.DataFrame(data)
df
```

```python
# 중복된 항목을 처리하기 위해 중복 행을 합칩니다.
df = df.groupby(['사용자 ID', 'ETF'])['구매'].size().reset_index()
# .max()
# .reset_index()
df
```

```python
# 사용자-ETF 행렬 생성
user_item_matrix = df.pivot(index='사용자 ID', columns='ETF', values='구매').fillna(0)
user_item_matrix
```

```python
# 사용자 유사성 계산 (예시: 코사인 유사성 사용)
from sklearn.metrics.pairwise import cosine_similarity

user_similarity = cosine_similarity(user_item_matrix)
user_similarity
```

```python
# 추천 함수 정의
def get_recommendations(user_item_matrix, user_similarity, user):
    # 사용자가 이미 구매한 ETF 제외
    user_purchases = user_item_matrix.loc[user]
    purchased_etfs = user_purchases[user_purchases > 0].index
    
    # 다른 사용자와의 유사도를 이용하여 추천 ETF 계산
    similar_users = user_similarity[int(user)]
#     user_similarity[int(user)]

    recommendations = []
    
    for similar_user, similarity_score in enumerate(similar_users):
        if similarity_score > 0:
            similar_user_purchases = user_item_matrix.iloc[similar_user]
            new_recommendations = similar_user_purchases[similar_user_purchases > 0].index
            for item in new_recommendations:
                if item not in purchased_etfs:
                    recommendations.append(item)
    
    return recommendations
```

```python
# 추천 받을 사용자 지정
target_user = 2

# 추천 목록 가져오기
recommendations = get_recommendations(user_item_matrix, user_similarity, target_user)

# 사용자에게 추천된 ETF 목록 출력
print(f"{target_user}에게 추천된 ETF 목록:")
print(recommendations)
```
